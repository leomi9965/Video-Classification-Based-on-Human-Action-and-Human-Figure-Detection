```python
IMG_SIZE = 224


def crop_center_square(frame):
   #
   #the function takes a numpy array frame as input, which is assumed to be an image or a frame from a video. 
   #It first finds the dimensions of the frame (height and width) using the shape attribute of the array. 
   #Then, it calculates the minimum dimension of the frame (either height or width), 
   #and finds the starting x and y coordinates for a square crop at the center of the frame.
   #Finally, it returns the cropped square.
   #
    y, x = frame.shape[0:2]
    min_dim = min(y, x)
    start_x = (x // 2) - (min_dim // 2)
    start_y = (y // 2) - (min_dim // 2)
    return frame[start_y : start_y + min_dim, start_x : start_x + min_dim]


def load_video(path, max_frames=0, resize=(IMG_SIZE, IMG_SIZE)):
  #
  #the function takes a file path path as input, which is assumed to be a video file. 
  #It opens the video file using OpenCV's VideoCapture function, and reads frames from the video using the read method.
   #For each frame, it crops a square from the center of the frame using the crop_center_square function, 
   #resizes the square to a specified size resize, and swaps the color channels from BGR to RGB 
   #(by slicing the last dimension of the array as [2, 1, 0]). The resulting frames are appended to a list frames, 
   #and the loop continues until either the end of the video is reached or the maximum number of frames max_frames is reached 
   #(if max_frames is greater than zero). Finally, the video file is released using the release method of VideoCapture, 
   #and the list of frames is converted to a numpy array and returned.
  #
    cap = cv2.VideoCapture(path)
    frames = []
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frame = crop_center_square(frame)
            frame = cv2.resize(frame, resize)
            frame = frame[:, :, [2, 1, 0]]
            frames.append(frame)

            if len(frames) == max_frames:
                break
    finally:
        cap.release()
    return np.array(frames)
```

```python
def build_feature_extractor():
    #
    #The function creates an instance of the InceptionV3 model, which is a pre-trained deep neural network for image classification. 
    #The weights parameter is set to "imagenet" to load the pre-trained weights for the model. 
    #The include_top parameter is set to False to exclude the top classification layer of the model, 
    #since we only want to use the features extracted from the convolutional layers of the network. 
    #The pooling parameter is set to "avg" to use average pooling after the convolutional layers. 
    #Finally, the input_shape parameter specifies the shape of the input images to the model.
    #

    feature_extractor = keras.applications.InceptionV3(
        weights="imagenet",
        include_top=False,
        pooling="avg",
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
    )

    #The function also creates an instance of the preprocess_input function from the InceptionV3 module, 
    #which is used to preprocess input images for the feature extractor model.
    preprocess_input = keras.applications.inception_v3.preprocess_input

    inputs = keras.Input((IMG_SIZE, IMG_SIZE, 3))
    preprocessed = preprocess_input(inputs)

    outputs = feature_extractor(preprocessed)
    return keras.Model(inputs, outputs, name="feature_extractor")


feature_extractor = build_feature_extractor()
```

```python
label_processor = keras.layers.StringLookup(num_oov_indices=0, vocabulary=np.unique(train_df["tag"]))
#The StringLookup layer is a Keras layer that maps string values to integer indices. 
#In this code, the num_oov_indices parameter is set to 0, which means that out-of-vocabulary (OOV) strings are not allowed. 
#The vocabulary parameter is set to a numpy array of unique labels in the training dataset, 
#which is obtained by calling the np.unique function on the "tag" column of the train_df DataFrame.
print(label_processor.get_vocabulary())
#he get_vocabulary method of the StringLookup layer is called to obtain the vocabulary of the layer. 
#This is a numpy array of strings that correspond to the integer indices used by the layer.


#Next, the labels variable is defined as the "tag" column of the df_train DataFrame, which presumably contains the labels for a training dataset. 
#The [..., None] indexing syntax is used to add an extra dimension to the labels array, since the StringLookup layer requires a 2D input.
labels = df_train["tag"].values
labels = label_processor(labels[..., None]).numpy()

labels
```

```python
#Define hyperparameters

IMG_SIZE = 224
BATCH_SIZE = 64
EPOCHS = 100

MAX_SEQ_LENGTH = 20 ###################################################
NUM_FEATURES = 2048
```

```python
def prepare_all_videos(df, root_dir):
    num_samples = len(df)
    video_paths = df["video_name"].values.tolist()
    
    ##take all classlabels from train_df column named 'tag' and store in labels
    labels = df["tag"].values
    
    #convert classlabels to label encoding
    labels = label_processor(labels[..., None]).numpy()

    # `frame_masks` and `frame_features` are what we will feed to our sequence model.
    # `frame_masks` will contain a bunch of booleans denoting if a timestep is
    # masked with padding or not.
    frame_masks = np.zeros(shape=(num_samples, MAX_SEQ_LENGTH), dtype="bool") # 145,20
    frame_features = np.zeros(shape=(num_samples, MAX_SEQ_LENGTH, NUM_FEATURES), dtype="float32") #145,20,2048

    # For each video.
    for idx, path in enumerate(video_paths):
        # Gather all its frames and add a batch dimension.
        frames = load_video(os.path.join(root_dir, path))
        frames = frames[None, ...]

        # Initialize placeholders to store the masks and features of the current video.
        temp_frame_mask = np.zeros(shape=(1, MAX_SEQ_LENGTH,), dtype="bool")
        temp_frame_features = np.zeros(
            shape=(1, MAX_SEQ_LENGTH, NUM_FEATURES), dtype="float32"
        )

        # Extract features from the frames of the current video.
        for i, batch in enumerate(frames):
            video_length = batch.shape[0]
            length = min(MAX_SEQ_LENGTH, video_length)
            for j in range(length):
                temp_frame_features[i, j, :] = feature_extractor.predict(
                    batch[None, j, :]
                )
            temp_frame_mask[i, :length] = 1  # 1 = not masked, 0 = masked

        frame_features[idx,] = temp_frame_features.squeeze()
        frame_masks[idx,] = temp_frame_mask.squeeze()
        print(idx)
    return (frame_features, frame_masks), labels


train_data, train_labels = prepare_all_videos(df_train, "train")
test_data, test_labels = prepare_all_videos(df_test, "test")

print(f"Frame features in train set: {train_data[0].shape}")
print(f"Frame masks in train set: {train_data[1].shape}")



print(f"train_labels in train set: {train_labels.shape}")

print(f"test_labels in train set: {test_labels.shape}")
```

```python
# Utility for our sequence model.
def get_sequence_model():
    class_vocab = label_processor.get_vocabulary()
 # The model takes two inputs: frame_features_input and mask_input. 
 # The frame_features_input input has shape (MAX_SEQ_LENGTH, NUM_FEATURES), 
 # where MAX_SEQ_LENGTH is the maximum length of a video sequence and NUM_FEATURES 
 # is the number of features extracted from each frame of the sequence. 
 # The mask_input input has shape (MAX_SEQ_LENGTH,) and is a binary mask indicating 
 # which elements of the frame_features_input input should be masked (ignored) during training.
    frame_features_input = keras.Input((MAX_SEQ_LENGTH, NUM_FEATURES))
    mask_input = keras.Input((MAX_SEQ_LENGTH,), dtype="bool")
    #The mask parameter is set to mask_input, which is a binary mask that indicates which elements of the input sequence should be masked (ignored) during training. 
    #This is useful when the input sequences have varying lengths, as the padding elements can be masked out during training.
    
    

    #The GRU layer is a type of recurrent neural network (RNN) layer that processes sequential data by maintaining a hidden state that is updated at each time step. 
    #The return_sequences parameter is set to True, which means that the layer will return the output at each time step instead of just the last time step. 
    #This is useful when subsequent layers require the entire sequence output instead of just a single output.
    x = keras.layers.GRU(16, return_sequences=True)(frame_features_input, mask=mask_input)
    x = keras.layers.GRU(8)(x)
    x = keras.layers.Dropout(0.4)(x)
    x = keras.layers.Dense(8, activation="relu")(x)
    output = keras.layers.Dense(len(class_vocab), activation="softmax")(x)

    rnn_model = keras.Model([frame_features_input, mask_input], output)

    rnn_model.compile(
        loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"]
    )
    return rnn_model

EPOCHS = 80
# Utility for running experiments.
def run_experiment():
    filepath = "./tmp/video_classifiers3"
    checkpoint = keras.callbacks.ModelCheckpoint(
        filepath, save_weights_only=True, save_best_only=True, verbose=1
    )

    seq_model = get_sequence_model()
    history = seq_model.fit(
        [train_data[0], train_data[1]],
        train_labels,
        validation_split=0.3,
        epochs=EPOCHS,
        callbacks=[checkpoint],
    )

    seq_model.load_weights(filepath)
    _, accuracy = seq_model.evaluate([test_data[0], test_data[1]], test_labels)
    print(f"Test accuracy: {round(accuracy * 100, 2)}%")

    return history, seq_model


_, sequence_model = run_experiment()
```

```python
def prepare_single_video(frames):
    frames = frames[None, ...]
    frame_mask = np.zeros(shape=(1, MAX_SEQ_LENGTH,), dtype="bool")
    frame_features = np.zeros(shape=(1, MAX_SEQ_LENGTH, NUM_FEATURES), dtype="float32")

    for i, batch in enumerate(frames):
        video_length = batch.shape[0]
        length = min(MAX_SEQ_LENGTH, video_length)
        for j in range(length):
            frame_features[i, j, :] = feature_extractor.predict(batch[None, j, :])
        frame_mask[i, :length] = 1  # 1 = not masked, 0 = masked

    return frame_features, frame_mask


def sequence_prediction(path):
    class_vocab = label_processor.get_vocabulary()

    frames = load_video(os.path.join("test", path))
    frame_features, frame_mask = prepare_single_video(frames)
    probabilities = sequence_model.predict([frame_features, frame_mask])[0]

    for i in np.argsort(probabilities)[::-1]:
        print(f"  {class_vocab[i]}: {probabilities[i] * 100:5.2f}%")
    return frames

test_video = np.random.choice(test_df["video_name"].values.tolist())
print(f"Test video path: {test_video}")

test_frames = sequence_prediction(test_video)
```
